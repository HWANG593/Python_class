```python
import pandas as pd

df = pd.DataFrame([[500, 450, 520, 610], 
                   [690, 700, 820, 900],
                   [1100, 1030, 1200, 1380],
                   [1500, 1650, 1700, 1850],
                   [1990, 2020, 2300, 2420],
                   [1020, 1600, 2200, 2550]],
                   index=[2015, 2016, 2017, 2018, 2019, 2020],
                   columns=['1분기', '2분기', '3분기', '4분기'])
# 1
display(df)
```

```python
# 2 

from matplotlib import font_manager, rc
font_path = "data/malgun.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()
print(font_name)
rc('font', family=font_name)

import matplotlib.pyplot as plt
import matplotlib
df.index = df.index.map(lambda x : str(x)+'년')

plt.figure()
plt.plot(df)
plt.legend(df.columns)
plt.title('시각화 과제1', color = 'purple')
plt.savefig('output/hw1.png')
```

```python
# 3

display(df.T)

df.T.plot(figsize = (10,6))
plt.title('시각화 과제2')
plt.savefig('output/hw2.png')
```

```python
# 4

colors = ['red', 'green', 'brown', 'purple']
explode = [0.05, 0.05, 0.05, 0.05]

plt.figure(figsize = (10,6))
plt.pie(df.sum(), 
        colors = colors, 
        explode = explode, 
        shadow = True, 
        autopct='%.1f%%',
        labels=df.columns)

plt.savefig('output/hw3.png')
```

```python
# 5

colors = ['red', 'green', 'brown', 'purple','yellow','pink']

plt.figure(figsize = (10,6))
plt.pie(df.T.sum(), 
        colors = colors, 
        autopct='%.1f%%',
        labels=df.T.columns.map(lambda x : str(x)+'년'))

plt.savefig('output/hw4.png')
```
