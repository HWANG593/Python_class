```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('./data/cctv_seoul.csv')

def ratio(a,b):
    return b/a*100

df['CCTV비율'] = df.apply(lambda x : ratio(x['인구수'], x['CCTV수']), axis=1)
df['노인비율'] = df.apply(lambda x : ratio(x['인구수'], x['고령자']), axis=1)
display(df)
```
